# WELCOME TO MY PALMER PENGUINS DATA VISUALIZATIONS
The following graphs are my practice of visualizing a famous dataset, Palmer Penguins using R programming. 

**Note:**
There are different ways to visualize the dataset but I will choose the 4 ways of visualizing below based on what I have learnt during the Google Data Analytics Certificate course.

```r
library(tidyverse)
library(palmerpenguins)
glimpse(penguins)

# Different ways to visualize the data
ggplot(data=penguins,aes(x=flipper_length_mm,y=body_mass_g))+
  geom_point(color="purple")

ggplot(data=penguins,aes(x=flipper_length_mm,y=body_mass_g))+
  geom_point(aes(shape=species))

# Different ways to generate the same visualization
ggplot(data=penguins,aes(x=flipper_length_mm,y=body_mass_g))+
  geom_point(aes(color=species,
                 shape=species)) +
  facet_wrap(~sex)

penguins %>%
  drop_na(sex) %>%
  ggplot(data=penguins,mapping=aes(x=flipper_length_mm,y=body_mass_g))+
  geom_point(aes(color=species,
                 shape=species)) +
  facet_wrap(~sex)
```
